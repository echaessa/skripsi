```python
#Case folding
data['casefold_full_text'] = data['full_text'].str.lower()
```

```python
#Text cleaning
import re
import string

def clean_text(text):
  text = re.sub(r'http\S+', '', text)  # Hapus URLs
  text = re.sub(r'@\S+', '', text)  # Hapus mentions
  text = re.sub(r'#', '', text)  # Hapus hashtags
  text = re.sub(r'\d+', '', text) # Hapus angka
  text = text.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation))) # Ganti semua tanda baca dengan spasi
  text = re.sub(r'\s+', ' ', text)  # Ganti spasi berlebih dengan satu spasi
  text = text.strip()  # Hapus spasi di depan dan di belakang
  return text

data['cleaned_full_text'] = data['casefold_full_text'].apply(clean_text)
```

```python
#Stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

stemmer_factory = StemmerFactory()
stemmer = stemmer_factory.create_stemmer()

def stemmed_text(text):
    words = text.split()
    words = [stemmer.stem(word) for word in words]
    return " ".join(words)

data['stemmed_full_text'] = data['stopword_full_text'].apply(stemmed_text)
```
